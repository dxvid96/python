# Visualisation using matplotlib

```python
plt.scatter(data.elevation, data.slope, color = 'orange') # plotting scatter plot
plt.xlabel('Elevation') # Labelling the x axis
plt.ylabel('Slope') # Labelling the y axis
plt.title('Relationship between Elevation and Slope') # Naming the title of the chart
```

```python
# Measuring multiple relationships on a 2 dimensional scatter graph

plt.scatter(data.elevation, data.isothermality) # plotting a scatter graph
plt.scatter(data.elevation, data.cropland_cover) # plotting a scatter graph: with consistent x-axis
plt.scatter(data.elevation, data.wind) # plotting another scatter graph
# this allows measurements of multiple relationships against a specific variable
plt.xlabel('Elevation') # naming x axis
plt.ylabel('other variables') # naming y axis
plt.title('Relationship between Elevation and other variables') # Adding title of chart
plt.legend(['isothermality', 'cropland', 'wind']) # adding legend to graph to represent data

# Looking at the graph below one could surmise there is no inherent relationship between elevation and 
# the variables measured
```

```python
# using a bar chart to measure groups across continuous variables
plt.bar(bank_data.marital, bank_data.age, color='grey') # barplot checking the ages of customers in different groups
```

```python
sns.catplot(x='marital', y='age', data=bank_data, kind='boxen') # a better way to investigate this on a granular level 
# using seaborn package

# Now unlike the first barplot which does not really represent the data depictively properly
# We can see across the different groups of marital statuses the different age caucauses in the dataset
# For example the average age of a single person is approximately 32 whereas the married group and divorced have an average of around 40
```

```python
sns.catplot(x='marital', y='age', data=bank_data, kind='boxen', hue='education') # adding a hue to gain more insight into data

# adding a hue gives more insight into behaviour across the groups across ages, education levels across 
# marital groups seems somewhat proportionally balanced but with many in the divorced group around age 50
# many individuals either have primary or unknown education levels
#this is why using different packages can prove useful for investigating data 
```

```python
# doing a stackplot: case scenario - person x spends their 24 hours on the following activities across a week

days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
sleeping = [7, 8, 6, 11, 7, 5, 7]
eating = [2, 3, 4, 3, 2, 5, 5]
working = [7, 8, 7, 2, 2, 3, 6]
playing = [8, 5, 7, 8, 13, 11, 6]
```

```python
# do a stackplot of person x's activities across a typical week

plt.stackplot(days, sleeping, eating, working, playing, colors=['red','blue','green','orange'])
# selecting the  lists and choosing different colours for them 
plt.legend(['sleep', 'eating', 'working', 'playing']) # creating a legend so any user can interpret data
plt.xlabel('Weekday') # naming x axis
plt.ylabel('Hours') # naming y axis
plt.title('Hours on Activity per day') # naming the chart title
plt.ylim(0,24) # to limit activity to 24 hours


```

```python
# 3d visualisation using matplotlib

from mpl_toolkits.mplot3d import axes3d # importing library that allows 3d visualisation
```

```python
fig = plt.figure() # creating a figure 
ax = plt.axes(projection='3d') # specifying that the figure is 3-dimensional 
ax.scatter(data.elevation, data.cropland_cover, data.isothermality, color='blue') # selecting the variables
ax.set_xlabel('Elevation') # naming x axis
ax.set_ylabel('Cropland') # naming y axis
ax.set_zlabel('Isothermality') # naming z axis
ax.set_title('3d graph of variables') # title of graph
ax.set_xlim(0,3000) # limiting the x axis to 3000
# This can be used to identify multivariate patterns inherent within data
```

```python
# pie charts

labels = 'Python', 'C++', 'Ruby', 'Java' # choosing the labels for the pie chart
sizes = [215, 130, 245, 210] # detailing the size of the different labels
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] #using colours for the different labels 
explode = (0, 0, 0, 0)  # this is to make certain slices stand out
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90) # adding patches and text, start
# angle to show where the legend should be
plt.legend(patches, labels, loc="lower right") # using patches as the legend alongside labels
# the loc can be used to specify where the legend is. This nullifies the impact of the startangle


plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
```

```python
# scatter plot using seaborn

sns.relplot(x='Birthrate', y='Deathrate', hue='Region', data=countries)

# based on this data which region would you estimate has a low birth to death ratio?
```

```python
# a beautiful thing about relplot is the ability to change the type of plot using the kind parameter

sns.relplot(x='Population', y='GDP ($ per capita)', hue='Region', kind= 'line', data=countries[(countries.Population>10000000)])



# Although the plot is messy which region has a positive relationship between population and GDP per capita?
# Approximately at what population does Asia's GDP per capita begin to pick up after a flat period?

```
